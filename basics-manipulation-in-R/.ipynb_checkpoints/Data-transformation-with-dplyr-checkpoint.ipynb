### Prerequisites
**Take careful note of the conflicts message that’s printed when you load the tidyverse.** 

It tells you that dplyr overwrites some functions in base R. If you want to use the base version of these functions after loading dplyr, you’ll need to use their full names: stats::filter() and stats::lag().

```r
library(nycflights13)
library(tidyverse)
```

```r
print(flights)
```

```r
print(filter(flights, month == 1, day == 1))
jan1 <- filter(flights, month == 1, day == 1)
```

```r
# The following code finds all flights that departed in November or December 
print(filter(flights, month == 11 | month == 12))
nov_dec <- filter(flights, month %in% c(11, 12))
```

```r
# find flights that weren’t delayed (on arrival or departure) 
# by more than two hours, you could use either of the following two filters:
print(filter(flights, arr_delay <= 120, dep_delay <= 120))
print(filter(flights, !(arr_delay > 120 | dep_delay > 120)))
```

```r
# filter will ignore FALSE condition and NA values unless being asked explicitly
df <- tibble(x = c(1, NA, 3))
print(filter(df, x > 1))
print(filter(df, is.na(x) | x > 1))
```

```r
# Departed in summer (July, August, and September)
filter(flights, between(month, 7,9))
```

```r
print(filter(flights, is.na(dep_time)))
```

```r
print(arrange(flights, desc(dep_delay)))
```

```r
print(select(flights, year, month, day))
```

```r
print(select(flights, -(year:day)))
```

```r
# When using rename function ,make sure to put the new_name = old_name
rename(flights, tail_num = tailnum)
```

```r
# Another option is to use select() in conjunction with the everything() helper. 
# This is useful if you have a handful of variables you’d like to move to the start of the data frame.
print(select(flights, time_hour, air_time, everything()))
```

```r
print(select(flights, contains("TIME")))
```

```r
flights_sml <- select(flights, 
                      year:day, 
                      ends_with("delay"), 
                      distance, 
                      air_time)
print(flights_sml)
```

```r
print(mutate(flights_sml,
       gain = dep_delay - arr_delay,
       speed = distance / air_time * 60
      ))
# Note that you can refer to columns that you’ve just created
print(mutate(flights_sml,
       gain = dep_delay - arr_delay,
       hours = air_time / 60,
       gain_per_hour = gain / hours
      ))
```

```r
# If you only want to keep the new variables, use transmute():
print(transmute(flights,
          gain = dep_delay - arr_delay,
          hours = air_time / 60,
          gain_per_hour = gain / hours
         ))
```

```r
by_day <- group_by(flights, year, month, day)
print(summarise(by_day, delay = mean(dep_delay, na.rm = TRUE)))
```

```r
by_dest <- group_by(flights, dest)
delay <- summarise(by_dest,
                   count = n(),
                   dist = mean(distance, na.rm = TRUE),
                   delay = mean(arr_delay, na.rm = TRUE)
                  )
delay <- filter(delay, count > 20, dest != "HNL")
print(delay)
```

```r
delays <- flights %>% 
          group_by(dest) %>%
          summarise(count = n(),
                    dist = mean(distance, na.rm = TRUE),
                    delay = mean(arr_delay, na.rm = TRUE)) %>%
          filter(count > 20, dest != "HNL")
print(delays)        
```
